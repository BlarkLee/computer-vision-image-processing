```python
import numpy as np
from scipy.misc import imread
import matplotlib.pyplot as plt

# Rotate image (img) by 90 anticlockwise
def rotate90(img):
    return np.flip(img,1).swapaxes(1,0)#Flip an array vertically (axis=0).
                                      #Flip an array horizontally (axis=1). 

def rotate180(img):
    return np.flip(np.flip(img,1),0)

def rotate270(img):
    return np.flip(img,0).swapaxes(1,0)

# Roate image (img) by an angle (ang) in anticlockwise direction
# Angle is assumed to be divisible by 90 but may be negative
def rotate(img, ang):
    assert ang%90==0
    num=ang/90%4
    if num==1:
        return rotate90(img)
    if num==2:
        return rotate180(img)
    if num==3:
        return rotate270(img)
    if num==4:
        return img
    
           
#Import image here
pic=imread('pepsi.jpg')
pic_90=rotate(pic,90)
pic_180=rotate(pic,180)
pic_270=rotate(pic,270)

#Sample call
pic_90=rotate(pic,90)
pic_180=rotate(pic,180)
pic_270=rotate(pic,270)
#Plotting code below
plt.subplot(2,2,1)
plt.imshow(pic)


plt.subplot(2,2,2)
plt.imshow(pic_90)


plt.subplot(2,2,3)
plt.imshow(pic_180)


plt.subplot(2,2,4)
plt.imshow(pic_270)
plt.show()
```
