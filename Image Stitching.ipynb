```python
import numpy as np
from scipy.misc import imread
import matplotlib.pyplot as plt

def merge(img1, img2, ncols):
    rows=img1.shape[0]
    columns=img1.shape[1]+img2.shape[1]-ncols
    img=np.zeros([rows, columns,3])
    for i in range (0,rows):
        for j in range(0,img1.shape[1]):
            img[i,j]=img1[i,j]
        for j in range(img1.shape[1],columns):
            img[i,j]=img2[i,j-img1.shape[1]+ncols]
    return img


left=imread('laptop_left.png')
right=imread('laptop_right.png')
print(left.shape) 
print(right.shape)
merge_pic=merge(left,right,15)
print(merge_pic.shape) 
plt.imshow(merge_pic)
plt.show()


```

```python
import numpy as np
from scipy.misc import imread
import matplotlib.pyplot as plt

def merge(img1,img2,ncols):
    raws=img1.shape[0]
    columns=img1.shape[1]+img2.shape[1]-ncols
    img=np.ones([raws,columns,3])
    img[:,:img1.shape[1],:]=img1[:,:,:]
    img[:,img1.shape[1]:,:]=img2[:,ncols:,:]
    return img
left=imread('laptop_left.png')
right=imread('laptop_right.png')
merge_pic=merge(left,right,15)
print(left.shape)
print(right.shape)
print(merge_pic.shape)
plt.imshow(merge_pic)
plt.show()
```
